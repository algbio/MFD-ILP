```python
import itertools
import math as m
import os 
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def read_input(graphfile):
    
    trip_data = open(graphfile,'r').read().split('\n')
    i = 0
    listOfGraphs = {}
    k = 0
    
    while(True):
        if "#" in trip_data[i]:
            i = i+1
            N = int(trip_data[i])
            edges = list()
            while(True):
                i = i+1;
                if "#" in trip_data[i]:
                    break;
                if "" == trip_data[i]:
                    break;
                line = trip_data[i].split(" ")
                edges.append((line[0],line[1],line[2]))
                if i >= len(trip_data)-1:
                    break;
            G = {'Nodes':N,'list of edges': edges}
            listOfGraphs[k] = G
            k +=1 
        if i >= len(trip_data)-1:
                    break; 

    return listOfGraphs;
    
```

```python
def flowMultipleDecomposition(data,K):
    
    #libraries
    import gurobipy as gp
    from gurobipy import GRB
    
    # calculate the minimal flow decomposition based on such graph
    V = data['vertices']
    E = data['edges']
    f = data['flows']
    W = data['maxFlow']
    M = 1e5
    S = data['sources']
    D = data['targets']
    AD_in = data['adj_in']
    AD_out = data['adj_out']
    N = data['newcycles']
    O = data['outcyclesdegrees']
    
    try:
        
        #create extra sets
        T = [(i,j,k) for (i,j) in E for k in range(0,K)]
        SC = [k for k in range(0,K)]
        SO = [(i,k) for i in range(0,len(N)) for k in range(0,K)]
        
        # Create a new model
        model = gp.Model("MFD")
        model.Params.LogToConsole = 0

        # Create variables
        x = model.addVars(T,vtype = GRB.BINARY, name="x")
        w = model.addVars(SC,vtype = GRB.INTEGER, name="w",lb=0)
        z = model.addVars(T,vtype = GRB.CONTINUOUS, name="z",lb=0)
        d = model.addVars(SO,vtype = GRB.BINARY,name="delta")
        
        model.setObjective(GRB.MINIMIZE)
       
        # flow conservation
        for k in range(0,K):
            for i in V:
                if i in S:
                    model.addConstr(sum(x[i,j,k] for j in V if i!=j and (i,j) in E)  == 1)  
                if i in D:
                    model.addConstr(sum(x[j,i,k] for j in V if i!=j and (j,i) in E) == 1)
                if i not in S and i not in D:
                    model.addConstr(sum(x[i,j,k] for j in V if i!=j and (i,j) in E) - sum(x[j,i,k] for j in V if i!=j and (j,i) in E) == 0)

        # flow balance
        for (i,j) in E:
            model.addConstr(f[i,j] == sum(z[i,j,k] for k in range(0,K)))

        # linearization
        for (i,j) in E:
            for k in range(0,K):
                model.addConstr(z[i,j,k] <= W*x[i,j,k])
                model.addConstr(w[k] - (1 - x[i,j,k])*W <= z[i,j,k])
                model.addConstr(z[i,j,k] <= w[k])
      

        # circuit elimination
        #model.addConstrs(gp.quicksum(x[i,j,k] for (i,j) in N[v]) >= len(N[v]) - M*(1 - d[v,k]) for k in range(0,K) for v in range(0,len(N)))
        #model.addConstrs(gp.quicksum(x[i,j,k] for (i,j) in N[v]) - len(N[v]) + 1 - W*d[v,k] <= 0 for k in range(0,K) for v in range(0,len(N)))
        #model.addConstrs(gp.quicksum(x[i,j,k] for (i,j) in O[v]) >= d[v,k] for k in range(0,K) for v in range(0,len(N)))
        
        # objective function
        model.optimize()
        
        # unpacked solutions
        w_sol = [0]*len(range(0,K))
        x_sol = {}
        paths = [list() for i in range(0,K)]
    
        if model.status == GRB.OPTIMAL:
            data['message'] = 'solved'
            data['runtime'] = model.Runtime;

            for v in model.getVars():
                if 'w' in v.VarName:
                    for k in range(0,K):
                        if str(k) in v.VarName:
                            w_sol[k] = v.x
                
                if 'x' in v.VarName:          
                    for (i,j,k) in T:
                        if str(i)+","+str(j)+","+str(k) in v.VarName:
                            x_sol[i,j,k] = v.x
                
                
            for(i,j,k) in T:
                if x_sol[i,j,k] == 1:
                    paths[k].append((i,j))
            
            data['weights'] = w_sol
            data['solution'] = paths
            data['x_sol'] = x_sol
        
        if model.status == GRB.INFEASIBLE:
            data['message'] = 'unsolved'
        
    except gp.GurobiError as e:
        print('Error code ' + str(e.errno) + ': ' + str(e))

    except AttributeError:
        print('Encountered an attribute error')
    
    return data;
    
```

```python
def checkCycles(data):
    
    solutionSet = data['solution']
    G = data['graph']
    V = data['vertices']
    S = {str(min(map(int,V)))}
    D = {str(max(map(int,V)))}
    
    # find cycles
    cyclesDetected = list()
    cyclesOutEdges = list()
    
    for i in range(len(solutionSet)):
        newTestGraph = nx.DiGraph()
        
        for v in solutionSet[i]:
            newTestGraph.add_edge(v[0],v[1],weight=1)
        
        try: 
            cycles = nx.find_cycle(newTestGraph)
        
        except nx.exception.NetworkXNoCycle as e:
            cycles = []
        
        if len(cycles) > 0:
            cyclesDetected.append(cycles)
    
    # find the nodes and edges from in and out a cycle
    for item in cyclesDetected:
        out_vertex = set(itertools.chain(*item))
        out_edges = list()
        for vertex in out_vertex:
            out_edges.extend(G.out_edges(vertex))
        
        out_of_cycle_edges = list()
        for (v1,v2) in out_edges:
            if ((str(v1),str(v2))) not in item:
                out_of_cycle_edges.append((str(v1),str(v2)))
        
        cyclesOutEdges.append(out_of_cycle_edges)
    
    return cyclesDetected,cyclesOutEdges
```

```python
def FD_Algorithm(data):
    
    listOfEdges = data['edges']
    solutionMap = data['graph']
    solutionSet = 0
    Kmin = data['minK']
    solutionWeights = 0

    for i in range(1,len(listOfEdges)+1):
        data = flowMultipleDecomposition(data,i)
        if data['message'] == "solved":
            solutionSet = data['solution']
            solutionWeights = data['weights']
            checkSCC(data,i);
            break;

    
    # display solutions
    colors = ['red','green','yellow','cyan','purple','violet']
    edges = list(solutionMap.edges())
    edges_colors = [0]*len(edges)
    edges_weights = [0]*len(edges)
    
    for i in range(0,len(solutionSet)):
        for k in range(0,len(edges)):
            edges_colors[k] = "beige"
        plt.figure()
        my_pos = nx.spring_layout(solutionMap, seed = 100)
        
        for j in solutionSet[i]:
            if j in edges:
                edges_colors[edges.index(j)] = "red"
                edges_weights[edges.index(j)] = 2
        
            
        print("W:",solutionWeights[i])
        nx.draw(solutionMap, pos = my_pos,edge_color=edges_colors,with_labels=True, node_color='lightblue', node_size=400, linewidths= edges_weights, font_size=15)
        plt.show()
        
    return data
```

```python
def SolveInstances(Graphs,outfile):
    
    fp = open(outfile,'w+')
    
    for i in range(0,1): 
        
        f = {}
        Edges = set()
        V = set()
        listOfEdges = Graphs[i]['list of edges']
        for k in range(0,len(listOfEdges)):
            (a,b,c) = (listOfEdges[k])
            Edges.add((a,b))
            V.add(a)
            V.add(b)
            f[a,b] = int(float(c))
            
        
        # creation of graphs
        G = nx.DiGraph()
        G.add_edges_from(Edges,weights = f)
        G.add_nodes_from(V)
        
        # creation of adjacent matrix
        AD_in = {}
        AD_out = {}
        
        for v in V:
            setAdj = set()
            for (i,j) in list(G.out_edges(v)):
                if i != v:
                    setAdj.add(i)
                if j != v:
                    setAdj.add(j)
            
            AD_out[v] = list(setAdj)
            
            setAdj = set()
            for (i,j) in list(G.in_edges(v)):
                if i != v:
                    setAdj.add(i)
                if j != v:
                    setAdj.add(j)
            
            AD_in[v] = list(setAdj)
        
        
        # calculating source, sinks and max flows
        S = [x for x in G.nodes() if G.out_degree(x)>=1 and G.in_degree(x)==0]
        D = [x for x in G.nodes() if G.out_degree(x)==0 and G.in_degree(x)>=1]
        maxW = max(f.values())
        
        
        # definition of data
        
        data = {'edges' : Edges,
                'flows' : f,
                'vertices' : V,
                'graph' : G,
                'Kmax' : len(Edges),
                'weights' : {},
                'sources': S,
                'targets': D,
                'message': {},
                'solution': 0,
                'maxFlow': maxW,
                'adj_in': AD_in,
                'adj_out': AD_out,
                'minK': 2,
                'runtime': 0,
                'newcycles': {},
                'x_sol':{},
                'outcyclesdegrees': {}
        }
        
        data = FD_Algorithm(data)        
        
        print(len(data['weights']),data['runtime'],file=fp)
        
    return 0
```
